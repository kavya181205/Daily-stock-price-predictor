```python
import yfinance as yf
import pandas as pd
```

```python
tata = yf.download(tickers=['TATASTEEL.BO'],period='5y',interval='1d',auto_adjust=True,multi_level_index=False)
reliance = yf.download(tickers='RELCHEMQ.BO',period='5y',interval='1d',auto_adjust=True,multi_level_index=False)
mrf = yf.download(tickers='MRF.BO',period='5y',interval='1d',auto_adjust=True,multi_level_index=False)
```

```python
tata.info(),reliance.info(),mrf.info()
```

```python
tata.reset_index(inplace=True)
reliance.reset_index(inplace=True)
mrf.reset_index(inplace=True)
```

```python
tata.drop(columns=['Volume'],inplace = True)
reliance.drop(columns=['Volume'],inplace = True)
mrf.drop(columns=['Volume'],inplace = True)
```

```python
tata.dropna(inplace=True)
reliance.dropna(inplace=True)
mrf.dropna(inplace=True)
```

```python
tata.info(),reliance.info(),mrf.info()
```

```python
tata['company'] = 'tata'
reliance['company'] = 'reliance'
mrf['company'] = 'mrf'
```

```python
df = pd.concat([tata,reliance,mrf])
```

```python
df.info()
```

```python
df.sort_values(by='Date',inplace=True)
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/final_destination.csv')
```
